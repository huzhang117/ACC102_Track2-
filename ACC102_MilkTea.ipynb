```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Milk_Tea.csv")
```

```python
df_clean = df.dropna()
df_clean = df_clean[df_clean["Brand"].str.contains("^[a-zA-Z &_.]+$",na=False)]
df_clean = df_clean.sort_values(by="Store_Count",ascending=False)
```

```python
plt.bar(df_clean["Brand"],df_clean["Store_Count"])
plt.title("Store Count Comparison")
plt.xticks(rotation=60)
plt.show()
```

```python
plt.bar(df_clean["Brand"],df_clean["Average_Price"],color="red")
plt.title("Average Price Comparison")
plt.xticks(rotation=60)
plt.show()
```

```python
plt.scatter(df_clean["Store_Count"],df_clean["City_Cover"])
plt.title("Store Count vs City Coverage")
plt.grid(True)
plt.show()
```
